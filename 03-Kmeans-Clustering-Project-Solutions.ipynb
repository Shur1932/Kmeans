## Подготовка данных и модель данных


### Отсутствующие данные (Missing Data)

**Строю отчёт - сколько строк имеют отсутствующие значения в той или иной колонке.**

```python
df.isnull().sum()
```

```python
df[df['Agriculture'].isnull()]['Country']
```

**Можно заметить следующее: большинство этих стран являются маленькими островами, кроме Гринландии и Западной Сахары. Заменяю для этих стран отсутствующие значения нулями, потому что это очень маленькие страны. Получилось 15 таких стран.**

    

```python
# Обнуляем неопределённые значения для маленьких островов
df[df['Agriculture'].isnull()] = df[df['Agriculture'].isnull()].fillna(0)
```

**Задание: Проверяю заново, в каких колонках сколько строк имеют неопределённые значения:**

```python
df.isnull().sum()
```

**Колонка "climate" пустая для нескольких стран. Но для всех стран заполнена колонка Region. Пользуясь этим, заменяю отсутствующие значения колонки Climate на средние значения этой колонки, где усреднение выполняется для соответствующего региона.**


```python
df['Climate'] = df['Climate'].fillna(df.groupby('Region')['Climate'].transform('mean'))
```

**Задание: Проверьте заново, сколько теперь отсутствующих значений.**

```python
df.isnull().sum()
```

**Для колонки "Literacy" применяю такой же метод, что и ранее для колонки Climate. Заменяю отсутствующие значения колонки Literacy % на средние значения этой колонки для соответствующего региона.**

```python
df[df['Literacy (%)'].isnull()]
```

```python
df['Literacy (%)'] = df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
```

**Задание: Проверяю заново, сколько теперь отсутствующих значений.**

```python
df.isnull().sum()
```

**Теперь данные отсутствуют только для нескольких стран. Удаляю такие страны**

```python
df = df.dropna()
```

## Подготовка данных с признаками

**Подготавливаю данные для кластеризации. Колонка Country является уникальным идентификатором, поэтому эта колонка не будет полезной для кластеризации, поскольку она содержит уникальные значения для каждой строки. Поэтому удаляю ее при создании фрейма Х с признаками.**

```python
X = df.drop("Country",axis=1)
```

**Создаю набор признаков X. Колонка Region всё ещё является категориальной, поэтому для этой колонки срздаю дамми-переменные с помощью Pandas.**

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

### Масштабирование данных

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

### Создание и обучение модели Kmeans




**Напишу цикл for для создания и обучения нескольких моделей KMeans, проверяя количество кластеров K от 2 до 30. На каждом шаге цикла сохраню сумму квадратов расстояний от точек до центров кластеров (SSD - Sum of Squared Distances) для каждого значения K. Далее нанесу эти данные на график, изображая метрику SSD для различных значений К.**

```python
from sklearn.cluster import KMeans
```

```python
ssd = []

for k in range(2,30):
    
    model = KMeans(n_clusters=k)
    
    
    model.fit(scaled_X)
    
    # Сумма квадратов расстояний (SSD - Sum of squared distances) от точек до центров кластеров
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,30),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```

Варианты K=3 или K=15 являются разумными вариантами

# Интерпретация модели


#### Выбираем K=3

**Можно сказать, что в точке K=3 падение метрики SSD более пологое (хотя метрика продолжает уменьшаться после этой точки). Как может выглядеть картина для K=3? Посмотрим, какие признаки являются важными в случае 3 кластеров**

```python
model = KMeans(n_clusters=3)
model.fit(scaled_X)
```

```python
model.labels_
```

```python
X['K=3 Clusters'] = model.labels_
```

```python
X.corr()['K=3 Clusters'].sort_values()
```
